# Load X-ray Images
Load the X-ray images from the specified directory or file paths.

```python
import os
import cv2
import matplotlib.pyplot as plt


def load_xray_images(directory):
    """
    Load X-ray images from the specified directory.

    Parameters:
    directory (str): The path to the directory containing X-ray images.

    Returns:
    list: A list of loaded X-ray images.
    """
    images = []
    for filename in os.listdir(directory):
        if (
            filename.endswith(".png")
            or filename.endswith(".jpg")
            or filename.endswith(".jpeg")
            or filename.endswith(".tiff")
        ):
            img_path = os.path.join(directory, filename)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                images.append(img)
    return images


# Example usage
directory = "./images"  # Replace with the actual path to your X-ray images
xray_images = load_xray_images(directory)

# Display the first loaded image
if xray_images:
    plt.imshow(xray_images[0], cmap="gray")
    plt.title("First Loaded X-ray Image")
    plt.axis("off")
    plt.show()
```

# Preprocess Images
Convert the images to grayscale if they are not already, and resize them to a standard size.

```python
# Preprocess Images

# Function to preprocess images: convert to grayscale if not already, and resize to a standard size
def preprocess_images(images):
    """
    Preprocess the X-ray images by converting them to grayscale and resizing them to a standard size.
    
    Parameters:
    images (list): A list of X-ray images.
    target_size (tuple): The target size for resizing the images (width, height).
    
    Returns:
    list: A list of preprocessed X-ray images.
    """
    preprocessed_images = []
    for img in images:
        # Check if the image is already in grayscale
        if len(img.shape) == 3 and img.shape[2] == 3:
            img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
        preprocessed_images.append(img)
    
    return preprocessed_images

# Example usage
preprocessed_xray_images = preprocess_images(xray_images)

# Display the first preprocessed image
if preprocessed_xray_images:
    plt.imshow(preprocessed_xray_images[0], cmap='gray')
    plt.title('First Preprocessed X-ray Image')
    plt.axis('off')
    plt.show()
```

# Apply Histogram Equalization
Use histogram equalization to adjust the contrast of the images so that they have similar gray levels.

```python
# Apply Histogram Equalization

def apply_histogram_equalization(images):
    """
    Apply histogram equalization to adjust the contrast of the X-ray images.
    
    Parameters:
    images (list): A list of X-ray images.
    
    Returns:
    list: A list of X-ray images with equalized histograms.
    """
    equalized_images = []
    for img in images:
        equalized_img = cv2.equalizeHist(img)
        equalized_images.append(equalized_img)
    
    return equalized_images

# Example usage
equalized_xray_images = apply_histogram_equalization(preprocessed_xray_images)

# Display the first equalized image
if equalized_xray_images:
    plt.imshow(equalized_xray_images[0], cmap='gray')
    plt.title('First Equalized X-ray Image')
    plt.axis('off')
    plt.show()
```

```python
# Apply CLAHE (Contrast Limited Adaptive Histogram Equalization)
def apply_clahe(images, clip_limit=2.0, tile_grid_size=(8, 8)):
    """
    Apply CLAHE to adjust the contrast of the X-ray images.

    Parameters:
    images (list): A list of X-ray images.
    clip_limit (float): Threshold for contrast limiting.
    tile_grid_size (tuple): Size of grid for histogram equalization.

    Returns:
    list: A list of X-ray images with CLAHE applied.
    """
    clahe = cv2.createCLAHE(clipLimit=clip_limit, tileGridSize=tile_grid_size)
    clahe_images = []
    for img in images:
        clahe_img = clahe.apply(img)
        clahe_images.append(clahe_img)

    return clahe_images


# Example usage
clahe_xray_images = apply_clahe(preprocessed_xray_images)

# Display the first CLAHE image
if clahe_xray_images:
    plt.imshow(clahe_xray_images[0], cmap="gray")
    plt.title("First CLAHE X-ray Image")
    plt.axis("off")
    plt.show()
```

# Display Results
Display the original and processed images side by side for comparison.

```python
# Display Results


# Function to display original, histogram equalized, and CLAHE processed images closer together
def display_results(original_images, equalized_images, clahe_images, num_images=5):
    """
    Display the original, histogram equalized, and CLAHE processed X-ray images closer together for comparison.

    Parameters:
    original_images (list): A list of original X-ray images.
    equalized_images (list): A list of histogram equalized X-ray images.
    clahe_images (list): A list of CLAHE processed X-ray images.
    num_images (int): The number of images to display for comparison.
    """
    plt.figure(figsize=(15, 10))
    for i in range(
        min(num_images, len(original_images), len(equalized_images), len(clahe_images))
    ):
        # Display original image
        plt.subplot(3, num_images, i + 1)
        plt.imshow(original_images[i], cmap="gray")
        plt.title(f"Original Image {i+1}")
        plt.axis("off")

        # Display histogram equalized image
        plt.subplot(3, num_images, num_images + i + 1)
        plt.imshow(equalized_images[i], cmap="gray")
        plt.title(f"Equalized Image {i+1}")
        plt.axis("off")

        # Display CLAHE processed image
        plt.subplot(3, num_images, 2 * num_images + i + 1)
        plt.imshow(clahe_images[i], cmap="gray")
        plt.title(f"CLAHE Image {i+1}")
        plt.axis("off")

    plt.tight_layout()
    plt.show()


# Example usage
display_results(xray_images, equalized_xray_images, clahe_xray_images)
```

```python
%pip install opencv-python scikit-image
import cv2
import numpy as np
from skimage import exposure, filters, morphology


# Placeholder for FABEMD function
def fabemd(image, K=10):
    # This function should decompose the image into K BIMFs and a residual image
    # For simplicity, we return dummy data
    BIMFs = [image for _ in range(K)]
    BR = image
    return BIMFs, BR


# Placeholder for energy calculation
def calculate_energy(BIMF):
    # This function should calculate the energy of a BIMF
    return np.sum(BIMF**2)


# Placeholder for nonlinear filtering
def nonlinear_filtering(BIMF):
    # This function should apply nonlinear filtering to a BIMF
    return cv2.medianBlur(BIMF, 5)


# Placeholder for Multi-Objective Optimization (MOO)
def multi_objective_optimization(IC):
    # This function should perform MOO to find the best parameters
    # For simplicity, we return the input image
    return IC


# Step 1: Decompose the CXR image
def step1_fabemd(image, K=10):
    return fabemd(image, K)


# Step 2: Process the residual image with a Homomorphic Filter
def step2_hmf(BR):
    dft = cv2.dft(np.float32(BR), flags=cv2.DFT_COMPLEX_OUTPUT)
    dft_shift = np.fft.fftshift(dft)
    rows, cols = BR.shape
    crow, ccol = rows // 2, cols // 2
    mask = np.ones((rows, cols, 2), np.uint8)
    r = 30
    center = [crow, ccol]
    x, y = np.ogrid[:rows, :cols]
    mask_area = (x - center[0]) ** 2 + (y - center[1]) ** 2 <= r * r
    mask[mask_area] = 0
    fshift = dft_shift * mask
    f_ishift = np.fft.ifftshift(fshift)
    IHMF = cv2.idft(f_ishift)
    IHMF = cv2.magnitude(IHMF[:, :, 0], IHMF[:, :, 1])
    return IHMF


# Step 3: Calculate the energy of the BIMFs
def step3_energy_calculation(BIMFs):
    return [calculate_energy(BIMF) for BIMF in BIMFs]


# Step 4: Denoise the lowest energy components and reconstruct the image
def step4_nonlinear_filtering_and_reconstruction(BIMFs, IHMF, energies, R=3, alpha=0.5):
    sorted_indices = np.argsort(energies)
    denoised_BIMFs = [
        nonlinear_filtering(BIMFs[i]) if i in sorted_indices[:R] else BIMFs[i]
        for i in range(len(BIMFs))
    ]
    IL = sum(denoised_BIMFs[:R]) + alpha * sum(BIMFs[R:]) + IHMF
    return IL


# Step 5: Apply Gamma Correction
def step5_gamma_correction(IL, gamma=1.2):
    return exposure.adjust_gamma(IL, gamma)


# Step 6: Apply CLAHE
def step6_clahe(I_gamma):
    return exposure.equalize_adapthist(I_gamma, clip_limit=0.03)


# Step 7: Multi-Objective Optimization
def step7_moo(IC):
    return multi_objective_optimization(IC)


# Main function to run the PACE 2.0 pipeline
def pace2_pipeline(image_path):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Step 1: FABEMD
    K = 10
    BIMFs, BR = step1_fabemd(image, K)

    # Step 2: Homomorphic Filtering
    IHMF = step2_hmf(BR)

    # Step 3: Energy Calculation
    energies = step3_energy_calculation(BIMFs)

    # Step 4: Nonlinear Filtering and Reconstruction
    IL = step4_nonlinear_filtering_and_reconstruction(BIMFs, IHMF, energies)

    # Step 5: Gamma Correction
    I_gamma = step5_gamma_correction(IL)

    # Step 6: CLAHE
    IC = step6_clahe(I_gamma)

    # Step 7: Multi-Objective Optimization
    Iopt = step7_moo(IC)

    # Output the enhanced image
    cv2.imshow("Original Image", image)
    cv2.imshow("Enhanced Image", Iopt)
    cv2.waitKey(0)
    cv2.destroyAllWindows()


# Run the pipeline
pace2_pipeline("images\1-IMA-01B_Thorax_AP.tiff")
```

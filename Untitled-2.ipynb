# Import Required Libraries
Import the necessary libraries, including TensorFlow, NumPy, OpenCV, and Matplotlib.

```python
%pip install tensorflow opencv-python matplotlib numpy scikit-learn
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, UpSampling2D, InputLayer
from sklearn.model_selection import train_test_split
```

# Load X-ray Images
Load the X-ray images from the specified directory or file paths.

```python
def load_xray_images(directory):
    """
    Load X-ray images from the specified directory.

    Parameters:
    directory (str): The path to the directory containing X-ray images.

    Returns:
    list: A list of loaded X-ray images.
    """
    images = []
    for filename in os.listdir(directory):
        if filename.endswith((".png", ".jpg", ".jpeg", ".tiff")):
            img_path = os.path.join(directory, filename)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                images.append(img)
    return images

# Example usage
directory = "./images"  # Replace with the actual path to your X-ray images
xray_images = load_xray_images(directory)

# Display the first loaded image
if xray_images:
    plt.imshow(xray_images[0], cmap="gray")
    plt.title("First Loaded X-ray Image")
    plt.axis("off")
    plt.show()
```

# Preprocess Images
Convert the images to grayscale if they are not already, and resize them to a standard size.

```python
def preprocess_images(images, target_size=(256, 256)):
    """
    Preprocess the X-ray images by converting them to grayscale and resizing them to a standard size.
    
    Parameters:
    images (list): A list of X-ray images.
    target_size (tuple): The target size for resizing the images (width, height).
    
    Returns:
    list: A list of preprocessed X-ray images.
    """
    preprocessed_images = []
    for img in images:
        # Resize image
        img = cv2.resize(img, target_size)
        preprocessed_images.append(img)
    return preprocessed_images

# Example usage
preprocessed_xray_images = preprocess_images(xray_images)

# Display the first preprocessed image
if preprocessed_xray_images:
    plt.imshow(preprocessed_xray_images[0], cmap='gray')
    plt.title('First Preprocessed X-ray Image')
    plt.axis('off')
    plt.show()
```

# Build CNN Model
Build a Convolutional Neural Network (CNN) model for normalizing gray levels in X-ray images.

```python
def build_cnn_model():
    model = Sequential()
    model.add(InputLayer(input_shape=(256, 256, 1)))
    model.add(Conv2D(32, (3, 3), activation='relu', padding='same'))
    model.add(MaxPooling2D((2, 2), padding='same'))
    model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
    model.add(MaxPooling2D((2, 2), padding='same'))
    model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
    model.add(MaxPooling2D((2, 2), padding='same'))
    model.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
    model.add(UpSampling2D((2, 2)))
    model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
    model.add(UpSampling2D((2, 2)))
    model.add(Conv2D(32, (3, 3), activation='relu', padding='same'))
    model.add(UpSampling2D((2, 2)))
    model.add(Conv2D(1, (3, 3), activation='sigmoid', padding='same'))
    return model

model = build_cnn_model()
model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()
```

# Train CNN Model
Train the CNN model using the preprocessed X-ray images.

```python
# Prepare data for training
images = np.array(preprocessed_xray_images)
images = np.expand_dims(images, axis=-1)  # Add channel dimension

# Split data into training and validation sets
x_train, x_val = train_test_split(images, test_size=0.2, random_state=42)

# Train the model
history = model.fit(x_train, x_train, epochs=50, batch_size=16, validation_data=(x_val, x_val))
```

# Evaluate and Visualize Results
Evaluate the trained model and visualize the original and normalized images.

```python
# Evaluate the model
loss = model.evaluate(x_val, x_val)
print(f'Validation Loss: {loss}')

# Visualize results
def visualize_results(model, x_val, num_images=5):
    for i in range(num_images):
        original_image = x_val[i]
        normalized_image = model.predict(np.expand_dims(original_image, axis=0))
        normalized_image = np.squeeze(normalized_image, axis=0)

        plt.figure(figsize=(10, 4))
        plt.subplot(1, 2, 1)
        plt.title('Original Image')
        plt.imshow(original_image.squeeze(), cmap='gray')
        plt.subplot(1, 2, 2)
        plt.title('Normalized Image')
        plt.imshow(normalized_image.squeeze(), cmap='gray')
        plt.show()

# Visualize the first few results
visualize_results(model, x_val)
```
